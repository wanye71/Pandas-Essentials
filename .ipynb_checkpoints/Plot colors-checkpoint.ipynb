# Basic Plotting

```python
import pandas as pd
```

```python
odf = pd.read_csv('olympics.csv', skiprows=4)
odf.head()
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

## Plot types

### What were the different sports in the first olypics? Plot them using different graphs

```python
first_olympics = odf[odf.Edition == 1896]
first_olympics.head()
```

### Line plot

```python
# Get the Sport count for the Olympics in 1896
first_olympics.Sport.value_counts()
```

```python
# Get the Sport count for the Olympics in 1896 shown in a line plot
first_olympics.Sport.value_counts().plot(kind='line')
```

```python
# Add a semicolon to the end of the statent to suppress '<Axes: xlabel='Sport'>'
first_olympics.Sport.value_counts().plot(kind='line');
```

### Bar plot

```python
first_olympics.Sport.value_counts().plot(kind='bar');
```

### Horizontal Bar plot

```python
first_olympics.Sport.value_counts().plot(kind='barh');
```

### Pie chart

```python
first_olympics.Sport.value_counts().plot(kind='pie');
```
