# Indexing

```python
import pandas as pd
```

```python
odf = pd.read_csv('../olympics.csv', skiprows=4)
odf.head()
```

```python
# Set the new index to a variable 'athlete'
athlete = odf.set_index('Athlete')
athlete.head()
```

```python
athlete.reset_index(inplace=True)
```

## sort_index()

```python
athlete.head()
```

```python
athlete.set_index('Athlete', inplace=True)
athlete.head()
```

```python
athlete.sort_index(inplace=True)
athlete.head()
```

```python
athlete.sort_index(inplace=True, ascending=False)
athlete.head()
```
